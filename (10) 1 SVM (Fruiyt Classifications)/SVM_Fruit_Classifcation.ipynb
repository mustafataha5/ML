```python
import pandas as pd

df =pd.read_csv('apples_and_oranges.csv')
```

```python
df
```

```python
import matplotlib.pyplot as plt 
plt.xlabel('Weight')
plt.ylabel('Size')

d_orange = df.query('Class == "orange"')
d_apple = df.query('Class == "apple"')

# Plot 'orange' data
plt.scatter(d_orange['Weight'], d_orange['Size'], color='orange', marker='o', label='Orange')

# Plot 'apple' data
plt.scatter(d_apple['Weight'], d_apple['Size'], color='red', marker='x', label='Apple') # Using 'red' and 'x' for distinction

# Add a title and legend for clarity
plt.title('Weight vs. Size for Oranges and Apples')
plt.legend()

# Display the plot
plt.show()
```

```python
X = df.iloc[:,:-1]
Y = df.iloc[:,-1]
```

```python
from sklearn.preprocessing import LabelEncoder 

le = LabelEncoder() 
Y = le.fit_transform(Y)
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y, test_size=.2, random_state=0 )
```

```python
from sklearn.svm import SVC 
model = SVC(kernel='linear')
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn import metrics 

cf = metrics.confusion_matrix(y_test,y_pred)
print("confusion_matrix:")
print(cf)


accuracy = metrics.accuracy_score(y_test,y_pred) 
precision = metrics.precision_score(y_test,y_pred) 
recall = metrics.recall_score(y_test,y_pred) 
f1_score = metrics.f1_score(y_test,y_pred)  


print(f'accuracy : {accuracy}')
print(f'precision : {precision}')
print(f'recall : {recall}')
print(f'f1_score : {f1_score}')
```
