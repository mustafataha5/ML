```python
import pandas as pd 

df = pd.read_csv('Groceries.csv')
```

```python
df
```

```python
df= df.T
df
```

```python
transations = df.apply(lambda x : x.dropna().tolist())
transations
```

```python
transations = transations.tolist()
```

```python
transations
```

```python
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder() 
rows = te.fit_transform(transations)
```

```python
df_transtion = pd.DataFrame(rows, columns=te.columns_)
df_transtion
```

```python
from mlxtend.frequent_patterns import apriori 

feat_uresitemsets = apriori(df_transtion,min_support=.005,use_colnames=True)
feat_uresitemsets['length'] = feat_uresitemsets['itemsets'].apply(lambda x: len(x))
feat_uresitemsets
```

```python
from mlxtend.frequent_patterns import association_rules 
rules = association_rules(feat_uresitemsets,metric='confidence',min_threshold=0.55)
rules = rules.sort_values(['confidence'],ascending=[False])

rules.nunique()
```

```python
from matplotlib import pyplot as plt 
from pandas.plotting import parallel_coordinates 

rules['antecedent'] =  rules['antecedents'].apply(lambda x: ','.join(list(x)))
rules['consequent'] = rules['consequents'].apply(lambda x: ','.join(list(x)))

rules['rule_n'] = rules.index

coord_rules = rules[['antecedent','consequent','rule_n']]

plt.figure(figsize=(4,8))

parallel_coordinates(coord_rules,'rule_n')
plt.legend([])
plt.grid(True)
plt.show()
```

```python
import seaborn as sns 

matrix = rules.pivot(index='antecedent',columns='consequent', values='confidence')

matrix
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(matrix,annot=True)
plt.yticks(rotation=0)
plt.xticks(rotation=90)
plt.show()
```

```python
def predict(items,rules,max_result = 6 ): 
    preds = rules[rules['antecedents'].apply(lambda x: items.issubset(x))]
    preds = preds[['antecedents','consequents','confidence']]
    # print(preds)
    preds.sort_values('confidence',ascending=False)
    return preds[:max_result]


p = predict({'whole milk'},rules,5)
print(p)
```
