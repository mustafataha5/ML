```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('income.csv')
df
```

```python
X= df[['Age','Education']]
Y = df['VIP']
```

```python
dfVip = df[df['VIP']==1]
dfnotVip  = df[df['VIP']==0]
```

```python
plt.scatter(dfVip['Education'],dfVip['Age'],color='red',marker='X')
plt.scatter(dfnotVip['Education'],dfnotVip['Age'],color='green',marker='o')
plt.legend(['VIP','notVIP'])

plt.show()
```

```python
from sklearn.linear_model import LinearRegression 
model = LinearRegression()
model.fit(X,Y)

```

```python
print(f"Intercept: {model.intercept_}")
print(f"Coefficients: {model.coef_}")
```

```python
b0 = model.intercept_[0]
b1 = model.coef_[0][0]
b2 = model.coef_[0][1]
```

```python
import math 
def sigmoid(z):
    return 1/(1+math.pow(math.e,-z))

age = 17 
education =18

y_pred = b0 +b1*age+b2*education 

y_Sigmode = sigmoid(y_pred)

if y_Sigmode > .5 : 
    print('(VIP)')
else: 
    print('(Not VIP)')

```

```python
### instead of manual create sigmode and calculate we use logistcRegression  
from sklearn.linear_model import LogisticRegression 

model_1 = LogisticRegression() 
model_1.fit(X.values,Y)
```

```python
pred=model_1.predict([[60,12]])
```

```python
if pred> .5 : 
    print('(VIP)')
else: 
    print('(Not VIP)')
```
