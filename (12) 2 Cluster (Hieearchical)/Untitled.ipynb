```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import numpy as np 
```

```python
x  = np.array([[1,1],[2,3],[1,9],[7,3],[2,7],[7,9],[8,4],[3,8],[4,1]])
x_swapped_1 = x[:, [1, 0]]
```

```python
x[:]
```

```python
x_swapped_1[:]
```

```python
x = x_swapped_1[:]
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(x[:,0],x[:,1],c='r')
for i in range(x.shape[0]):
    plt.annotate(str(i),xy=[x[i,0],x[i,1]],xytext=(3,3),textcoords='offset points')

plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Scatter plot if X Y Data")
tick_locations = np.arange(0, 11, 1) # Goes up to but not including 11, so it includes 10

plt.xticks(tick_locations) # Apply to X axis
plt.yticks(tick_locations) # Apply to Y axis  
plt.grid()

plt.show()
    
```

```python
from scipy.cluster.hierarchy import dendrogram,linkage 


z1 = linkage(x,method='single',metric='euclidean')
z2 = linkage(x,method='complete',metric='euclidean')
z3 = linkage(x,method='average',metric='euclidean')
z4 = linkage(x,method='ward',metric='euclidean')
```

```python
plt.figure(figsize=(10, 7))
plt.subplot(2,2,1,)
dendrogram(z1)
plt.title('single')

plt.subplot(2,2,2,)
dendrogram(z2)
plt.title('complete')

plt.subplot(2,2,3,)
dendrogram(z3)
plt.title('average')

plt.subplot(2,2,4,)
dendrogram(z4)
plt.title('ward')
```

```python
from scipy.cluster.hierarchy import fcluster 

f1 =fcluster(z4,2,criterion='maxclust')
print(f'F1 ',f1)
```

```python
from sklearn.cluster import AgglomerativeClustering 
model = AgglomerativeClustering(n_clusters=2,linkage='ward')
```

```python
z=model.fit(x)
print(z.labels_)
print(z.labels_*-1+2)
```
